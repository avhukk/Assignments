#### 5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.

```python
import os 
import shutil 

def search_transfer(source, destination, file_ext): 
    source = os.path.abspath(source)
    destination = os.path.abspath(destination)
    try: 
        for foldername, subfolder, files in os.walk(source): 
            print(f'folder name is {foldername}', end = "\n\n\n")
            print(f'sub folders are {subfolder}', end = "\n\n\n")
            print(f'files/contents within the subfolder {files}', end = "\n\n\n")
            for file in files: 
                filename, extension = os.path.splitext(file)
                if extension in file_ext:
                    targetfile = foldername+os.path.sep+filename+extension
                    shutil.copy(targetfile, destination)
            print("files copied successfully")
    except Exceptions as e: 
        print(e)
```
